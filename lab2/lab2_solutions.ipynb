```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

 # Berkson's paradox

## Draw 1000 samples of $x$ and $y$ from uniform distribution and a) calculate Pearson correlation, b) Plot scatterplot

```python
x = np.random.uniform(0,1,1000)
y = np.random.uniform(0,1,1000)
```

```python
plt.scatter(x, y)
scipy.stats.pearsonr(x, y)[0]
```

## Condition dataset on the collider: $x$ + $y$ > 0.2  and calculate Pearson correlation, and a) calculate Pearson correlation, b) Plot scatterplot

```python
threshold = 0.2

new_x = x[x > threshold - y]
new_y = y[y > threshold - x]

#  # For plotting a calculation of r^2 the arrays must be the same length
    
no_samples = min(len(new_x), len(new_y))
new_x = new_x[:no_samples]
new_y = new_y[:no_samples]
```

```python
plt.scatter(new_x, new_y)
scipy.stats.pearsonr(new_x, new_y)[0]
```

## See how the correlation and the scatterplots change for $threshold$ in range(0.0, 2.0, 0.1). Contemplate.

```python
for thresh in np.arange(0.0, 2.0, 0.1):
    new_x = x[x > thresh - y]
    new_y = y[y > thresh - x]
    
    no_samples = min(len(new_x), len(new_y))
    new_x = new_x[:no_samples]
    new_y = new_y[:no_samples]
    
    plt.scatter(new_x, new_y)
    plt.show()
    print(thresh, scipy.stats.pearsonr(new_x, new_y)[0])
```
