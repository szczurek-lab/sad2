# d-separation

## Why is d-separation so useful?

D-separation is a powerful tool for determining whether two variables are independent of one another. When two variables are d-separated, they are said to be conditionally independent of one another. 

D-separation can be used to determine whether two variables are independent of one another. For example, we might want to know whether two variables, X and Y, are independent of one another. To do this, we would need to find a third variable, Z, that is d-separated from both X and Y. If we can find such a variable, then we can say that X and Y are conditionally independent of one another given Z.
